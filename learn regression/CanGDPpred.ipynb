```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model as model
reg=model.LinearRegression()
```

```python
df=pd.read_csv('percap.csv')
df
```

```python
plt.scatter(df.year,df.percap,marker='x',color='r')
plt.xlabel('year')
plt.ylabel('Per Capita Income in US$')
plt.show()
```

```python
reg.fit(df[['year']],df.percap)
```

```python
plt.scatter(df.year,df.percap,marker='x',color='r')
plt.plot(df.year,reg.predict(df[['year']]),color='b')
plt.xlabel('year')
plt.ylabel('Per Capita Income in US$')
plt.show()
```
