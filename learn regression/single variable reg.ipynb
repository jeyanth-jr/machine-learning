```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model as lm
model=lm.LinearRegression()
```

```python
df=pd.read_csv('data.csv')
df
```

```python
plt.scatter(df.area,df.price,marker='x',color='r')
plt.xlabel('area')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
model.fit(df[['area']],df.price)
```

```python
plt.scatter(df.area,df.price,marker='x',color='r')
plt.plot(df.area,model.predict(df[['area']]),color='b')
plt.xlabel('area')
plt.ylabel('price')
plt.legend()
plt.show()
```
